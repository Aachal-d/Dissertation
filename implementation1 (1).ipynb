# **Make a new dataframe with just the filtered activities and ratings for all users for that particular activity**

```python
new_df = user[['user','activity','rating']].copy()
```

```python
length = len(list2)
length_set = len(new_df)
new_new_df = pd.DataFrame({'user':[],'activity':[],'rating':[]})

dict_list = []

for i in range(0,length-1):
  for j in range(0,length_set-1):
    if new_df.activity[j] == list2[i]:
      row_dict = { 'user': new_df.user[j], 'activity' : new_df.activity[j], 'rating' : new_df.rating[j] }
      dict_list.append(row_dict)

new_new_df = pd.DataFrame.from_dict(dict_list)

print(new_new_df)
```

```python
user_ratings = new_new_df.pivot_table(index = {'user'}, columns = {'activity'}, values = 'rating')
user_ratings.head()
```

# **Use collaborative filtering**

```python
user_ratings = user_ratings.fillna(0)
user_ratings
```

```python
# Normalize user-item matrix
matrix_norm = user_ratings.subtract(user_ratings.mean(axis=1), axis = 'rows')
matrix_norm.head()
```

```python
# User similarity matrix using Pearson correlation
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
# Remove the person with user ID 12
user_similarity.drop(index=12, inplace=True)
user_similarity.head()
```

```python
 # User similarity threashold of 0.4 or more
# Get top 4 similar users
drop_user = 12
similar_users = user_similarity[user_similarity[drop_user]>0.4][drop_user].sort_values(ascending=False)[:4]
# Print out top 4 similar users
print(f'The similar users for user {drop_user} are', similar_users)
```

```python
# activities that similar users have tried and discard the activities that none of the similar users have tried
similar_user_activity = matrix_norm[matrix_norm.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_activity
```

```python
# A dictionary to store item scores
item_score = {}

for i in similar_user_activity.columns:
  activity_rating = similar_user_activity[i]
  total = 0
  count = 0
# score is the sum of user similarity score multiply by the activity rating then we add the score to the total score for the activity so far
  for j in similar_users.index:
    if pd.isna(activity_rating[j]) == False:
      score = similar_users[j] * activity_rating[j]
      total = total + score
      count = count + 1
# Get the average score for the item
  item_score[i] = total / count
    
# Convert dictionary to pandas dataframe where the activities are ranked by the score and print top 4 by score
item_score = pd.DataFrame(item_score.items(), columns=['activity', 'activity_score'])
ranked_item_score = item_score.sort_values( by = 'activity_score', ascending=False )
ranked_item_score.head(6)
```

```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import cross_validate
from surprise.accuracy import rmse

# Load data from a DataFrame
df = pd.read_csv('Book3.csv')

# Define the rating scale
reader = Reader(rating_scale=(1, 5))

# Load data into Surprise dataset
data = Dataset.load_from_df(df[['user', 'activity', 'rating']], reader)

# Define the SVD algorithm
algo = SVD()

# Evaluate the algorithm using RMSE metric
cv_results = cross_validate(algo, data, measures=['RMSE','MAE'], cv=5, verbose=True)

# Print the average RMSE score
print("Average RMSE score:", cv_results['test_rmse'].mean())
print("Average MAE score:", cv_results['test_mae'].mean())
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
ground_truth = {'activity' : {'Board games', 'Cooking', 'Walking', 'Reading', 'Puzzles', 'Craft','Table Tennis'} }
predicted_labels = {'activity' : {'Reading', 'Craft', 'Board games', 'Cooking', 'Puzzles', 'Walking'}}

ground_truth = [0, 1, 2, 3, 4, 5, 14]
predicted_labels = [3, 5, 0, 1, 4, 2]
```

```python
# Calculate the number of correctly recommended items
correct_recommendations = set(predicted_labels).intersection(set(ground_truth))
num_correct_recommendations = len(correct_recommendations)

# Calculate precision
precision = num_correct_recommendations / len(predicted_labels)

# Calculate recall
recall = num_correct_recommendations / len(ground_truth)

# Calculate F1-score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the evaluation metrics
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1_score)
```
