# **Using Content Based Filtering  **

```python
list1= []
count = 0
list2= []

if user.mental_illness[56] == 'yes':
  if user.illness_type[56] == 'anxiety':
    while count <24:
      list1.append(activity.cognitive[count])
      list1.append(activity.physical[count])
      list1.append(activity.social[count])
      list1.append(activity.cultural[count])
      if (list1 == [1,0,0,0]) | (list1 == [0,1,0,0]):
        list2.append(activity.activity[count])
        print(activity.activity[count])
      count = count+1 
      list1 = []
  elif user.illness_type[56] == 'depression':
    while count <24:
      list1.append(activity.cognitive[count])
      list1.append(activity.physical[count])
      list1.append(activity.social[count])
      list1.append(activity.cultural[count])
      if (list1 == [1,0,0,0]) | (list1 == [0,1,0,0]) | (list1 == [0,0,1,0]) | (list1 == [0,0,0,1]) :
        list2.append(activity.activity[count])
        print(activity.activity[count])
      count = count+1 
      list1 = []
  elif user.illness_type[56] == 'neurodevelopmental':
    while count <24:
      list1.append(activity.cognitive[count])
      list1.append(activity.physical[count])
      list1.append(activity.social[count])
      list1.append(activity.cultural[count])
      if list1 == [1,0,0,0]:
        list2.append(activity.activity[count])
        print(activity.activity[count])
      count = count+1 
      list1 = []
elif user.mental_illness[56] == 'no':
  while count <24:
      list1.append(activity.cognitive[count])
      list1.append(activity.physical[count])
      list1.append(activity.social[count])
      list1.append(activity.cultural[count])
      if (list1 == [1,0,0,0]) | (list1 == [0,1,0,0]) | (list1 == [0,0,1,0]) | (list1 == [0,0,0,1]):  
        list2.append(activity.activity[count])
        print(activity.activity[count])
      count = count+1 
      list1 = []
```

```python
new_df = activity[['activity','description']].copy()
```

```python
length = len(list2)
length_set = len(new_df)
new_new_df = pd.DataFrame({'activity':[],'description':[]})

dict_list = []

for i in range(0,length-1):
  for j in range(0,length_set-1):
    if new_df.activity[j] == list2[i]:
      row_dict = { 'activity': new_df.activity[j], 'description' : new_df.description[j] }
      dict_list.append(row_dict)

new_new_df = pd.DataFrame.from_dict(dict_list)

print(new_new_df)
```

```python
#fill in the blank descriptions with an empty string and form a tfidf matrix
tfidf = TfidfVectorizer(stop_words='english')
new_new_df['description'] = new_new_df['description'].fillna('')
tfidf_matrix = tfidf.fit_transform(new_new_df['description'])
```

```python
tfidf_matrix.shape
```

```python
cosine_similarity = linear_kernel( tfidf_matrix , tfidf_matrix )
```

```python
activity.index
```

```python
#we drop the duplicates and stores the index in a series (holds any data)
indices = pd.Series(activity.index, index=activity['activity']).drop_duplicates()
```

```python
#activity_x is the activity to which we want a similar activity recommended
#get_recommendation( activity_x , cosine_sim=cosine_similarity ):
index = indices['Football']
similarity_scores = list(enumerate(cosine_similarity[index]))
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
similarity_scores = similarity_scores[1:7]
activity_indices = [i[0] for i in similarity_scores]

second = []
for i in similarity_scores:
  second.append(i[1]) 

#iloc selects the row with index activity_index and returns the name of the similar activity calculated according to similarity scores as a newd dataframe
data = pd.DataFrame({'activity' : activity['activity'].iloc[activity_indices], 'scores' : second})

print(data)
```

# **Print similar activities for other activities**

```python
index = indices['Running']
similarity_scores = list(enumerate(cosine_similarity[index]))
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
similarity_scores = similarity_scores[1:7]
activity_indices = [i[0] for i in similarity_scores]

second = []
for i in similarity_scores:
  second.append(i[1]) 

#iloc selects the row with index activity_index and returns the name of the similar activity calculated according to similarity scores as a newd dataframe
data = pd.DataFrame({'activity' : activity['activity'].iloc[activity_indices], 'scores' : second})

print(data)
```

# **Evaluation**

```python
# doesn't make sense to calculate the mse and mae for content based filtering beacause we do not calculate the predicted rating for each of the activites.
```

```python
ground_truth = ['Basketball', 'Volleyball', 'Hockey', 'Tennis', 'Badminton']

# Get the recommended activities
recommended_activities = list(data['activity'])

# Calculate the number of correctly recommended items
correct_recommendations = set(recommended_activities).intersection(set(ground_truth))
num_correct_recommendations = len(correct_recommendations)

# Calculate precision
precision = num_correct_recommendations / len(recommended_activities)

# Calculate recall
recall = num_correct_recommendations / len(ground_truth)

# Calculate F1-score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print the evaluation metrics
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1_score)
```
